**ФУНКЦИИ ДВУХ ПЕРЕМЕННЫХ.**
График такой функции представляет из себя поверхность.

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

def f(x, y): 
    return (2*x +y - 8)**2 - 3*x*y

fig=plt.figure()
ax = fig.gca(projection='3d')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)

X, Y = np.meshgrid(X, Y)
Z = f(X,Y)

#ax=Axes3D(fig)
ax.plot_surface(X,Y,Z,color='blue') 

plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

def f(x, y): 
    return x + 24*y - 11

x = np.linspace(1, 10, 10) 
y = np.linspace(1, 10, 10)

fig = plt.figure()
ax = plt.axes(projection='3d')

X, Y = np.meshgrid(x, y) 
Z = f(X, Y) 

ax.plot_wireframe(X, Y, Z, color ='blue') 

plt.show()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

def f(x, y): 
    return x ** 2 + 2 * y ** 2 

x = np.linspace(1, 10, 10) 
y = np.linspace(1, 10, 10)

fig = plt.figure()
ax = plt.axes(projection='3d')

X, Y = np.meshgrid(x, y) 
Z = f(X, Y) 

ax.plot_wireframe(X, Y, Z, color ='blue') 

plt.show()
```
