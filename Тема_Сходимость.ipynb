**ПОСТРОЕНИЕ ГРАФИКА ФУНКЦИИ.**
Пример ниже иллюстрирует основные средства построения графиков с использование библиотеки matplotlib как графической среды и библиотеки numpy как средства работы с фунциями.

```python
#functions
import numpy as np

#graphics from matplotlib library as interface pyplot
import matplotlib.pyplot as plt

#range of values: from 0 to 4pi with number of points 100
x = np.linspace(0, 4 * np.pi, 100)


plt.plot(x, np.sin(x), 'r-')# r- red colour and unbroken line
plt.plot(x, np.cos(x), 'b--')# b-- blue colour and dashed line
plt.grid()#grid showing
plt.show()#plot showing
```

**ИЛЛЮСТРАЦИЯ ПРЕДЕЛОВ ФУНКЦИЙ.**
Если функция непрерывна, то пределом функции в окрестности некоторой точки является само значение функции. В местах разрыва могут быть различные ситуации: существование одного предела, двух разные пределов слева и справа от рассматриваемой точки, или пределы могут быть бесконечностями.

```python
#range of values: from -10pi to 10pi with number of points 100
x = np.linspace(-10 * np.pi, 10 * np.pi)
plt.plot(x, np.arctan(x), 'b-')

#arrow image
plt.arrow(15, 1.3, 10, 0,
          head_length = 5,
          head_width = 0.15,
          width = 0.05,
          color = 'r')

plt.annotate('y -> Pi/2', xy = (20, 1.0))

plt.arrow(-15, -1.3, -10, 0,
          head_length = 5,
          head_width = 0.15,
          width = 0.05,
          color = 'r')

plt.annotate('y -> - Pi/2', xy = (-27, -1.1))

plt.grid()#grid showing
plt.show()#plot showing
```

```python
#range of values: from -5 to 5 with number of points 100
x = np.linspace(-2, 6, 100)

x[(x>1.9) & (x<2.0)] = np.nan #correction of the break point

#image of the asymptote
plt.arrow(2,0,0,1,
          color = 'r',
          width = 0.015,
          linestyle='dashed')

#arrow image
plt.arrow(0, 1.0, 1, 0,
          head_length = 0.5,
          head_width = 0.05,
          width = 0.02,
          color = 'r')

plt.annotate('y -> 1', xy = (0, 0.9))

plt.arrow(4, 0.0, -1, 0,
          head_length = 0.5,
          head_width = 0.05,
          width = 0.02,
          color = 'r')

plt.annotate('y -> 0', xy = (3, 0.1))

plt.plot(x,(1/(1 + np.exp(1 / (x - 2)))),'b-')
plt.grid()#grid showing
plt.show()#plot showing
```

```python
#definition of an user fucntion
def func(x):
  if 0<=x<6:
    return 2-(1/(6-x))
  elif 6<=x<10:
    return x - 4

x = np.linspace(-1, 11, 100)
x[(x>5.95) & (x<6.1)] = np.nan #correction of the break point
#image of the asymptote
plt.arrow(6,6,0,-10,
          color = 'r',
          width = 0.008,
          linestyle='dashed')

#arrow image
plt.arrow(4, -3.5, 1, 0,
          head_length = 0.5,
          head_width = 0.4,
          width = 0.15,
          color = 'r')

plt.annotate('y -> inf', xy = (4, -3.0))

plt.arrow(8, 2.0, -1, 0,
          head_length = 0.5,
          head_width = 0.4,
          width = 0.15,
          color = 'r')

plt.annotate('y -> 2', xy = (7, 1.3))

y = np.vectorize(func)

plt.plot(x,y(x),'b-')
plt.xlim([0, 10])#range for x
plt.ylim([-4, 6])#range for y
plt.grid()#grid showing()
plt.show()#plot showing
```

**ИЛЛЮСТРАЦИЯ РАЗРЫВОВ ФУНКЦИИ В НАБОРЕ ДАННЫХ.**
На практике одной из задач в обработке данных является нахождение функциональной зависимости, порождающей все значения параметров в наборе данных. Ниже проиллюстрирован пример, когда строится зависимость (синяя линия) в некотором наборе данных (черные точки), и эта зависимость является кусочно-гладкой функцией с разрывом. Таким образом, набор данных ниже некоторого значения характеризуется одной функцией, а выше этого значения - имеет совсем другое поведение.

```python
#dataset
x_data = list(range(20))
y_data = [0.9, 0.85, 1.9, 4.5, 6.27, 11.65, 12.3, 19, 23, 29.31, 20, 20.5, 22.78, 23, 24.2, 24.55, 25, 26.41, 28.6, 29]

#definition of an user fucntion
def func2(x):
  if 0<=x<10:
    return 0.388 * x**2
  elif 10<=x<20:
    return x + 10

x = np.linspace(0, 20, 100)
x[(x>9.88) & (x<10.1)] = np.nan #correction of the break point
#image of the asymptote
plt.arrow(10,0,0,40,
          color = 'r',
          width = 0.008,
          linestyle='dashed')

y = np.vectorize(func2)
plt.plot(x,y(x),'b-')


#arrow image
plt.arrow(15, 19.5, -3, 0,
          head_length = 1,
          head_width = 2,
          width = 0.5,
          color = 'r')

plt.annotate('y -> 20', xy = (12, 17.2))


plt.arrow(5, 30, 3, 0,
          head_length = 1,
          head_width = 2,
          width = 0.5,
          color = 'r')

plt.annotate('y -> 30', xy = (5, 31))


plt.scatter(x_data, y_data, c='black', label='data')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
