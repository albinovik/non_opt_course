**Скалярная функция. Производная по направлению. Градиент.**


Когда набору координат в соответствие ставится число, то говорят, что задана скалярная функция. Можно проиллюстрировать эту ситуацию в электростатие:

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

def f(x, y): 
    return x * np.exp(-x**2 - y**2)

fig=plt.figure()
ax = fig.gca(projection='3d')
X = np.arange(-2, 2, 0.1)
Y = np.arange(-2, 2, 0.1)

X, Y = np.meshgrid(X, Y)
Z = f(X,Y)

#ax=Axes3D(fig)
ax.plot_surface(X,Y,Z,color='green') 

plt.show()
```
