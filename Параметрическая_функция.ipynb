**ПАТАМЕТРИЧЕСКАЯ ФУНКЦИЯ И ЕЁ ПРОИЗВОДНАЯ.**
Часто можно встретиться с параметрическим заданием функций. Например, в физике траектория в виде спирали хорошо описывается через параметры.

```python
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(0,4*np.pi,100)
x=np.cos(t)
y=np.sin(t)

plt.plot(x, y, 'b-')# b- red colour and unbroken line
plt.grid()#grid showing
plt.show()#plot showing
```

```python
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(0,2*np.pi,100)
x=(t + 1)
y=t - 2

plt.plot(x, y, 'b-')# b- red colour and unbroken line
plt.grid()#grid showing
plt.show()#plot showing
```

```python
from mpl_toolkits.mplot3d import Axes3D

t=np.linspace(0,4*np.pi,100)
x=np.cos(t)
y=np.sin(t)
z=t/(4*np.pi)

fig=plt.figure()
ax=Axes3D(fig)
ax.plot(x,y,z,'b--')
```

Производная может быть найдена по известным формулам.

```python
import sympy as sym

x = sym.Symbol('x')
y = sym.Symbol('y')
t = sym.Symbol('t')

x=sym.cos(t)
y=sym.sin(t)

f1 = x.diff(t)
f2 = y.diff(t)
f3 = f2 / f1

f3
```
